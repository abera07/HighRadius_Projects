```
d2 = data.groupby('State').size().to_dict()
```

```
d3 = {k : d1[k]/d2[k] for k in d1.keys()&d2}
```

```
d4 = {k : 1-d3[k] for k in d3.keys()}
```

```
data['Class'] = data['State'].map(d4)
data['State'] = data['State'].map(d3)
```
